def make_change(cents: int):
  quarters = cents // 25  #  // does "integer division" (computes quotient)
  cents -= quarters * 25
  dimes = cents // 10
  cents -= dimes * 10
  nickels = cents // 5
  cents -= nickels * 5
  pennies = cents
  return [quarters, dimes, nickels, pennies]

*Python offers some nice shortcuts. For example, the following code:

```python
[quarters, dimes, nickels, pennies] = make_change(cents) # [shortcut]
```

is a shortcut for:

```python
change   = make_change(cents)
assert len(change) == 4
quarters = change[0]
dimes    = change[1]
nickels  = change[2]
pennies  = change[3]
```

and you don't even need the `[]`:

```python
quarters, dimes, nickels, pennies = make_change(cents) # shortcut
```

As we will learn later, just when you thought it couldn't get any shorter, the entire thing can be written as:

```python
test_change(cents, *make_change(cents))  # *wow!!!
```

The `*` here unpacks the list returned by `make_change` and passes them in order as arguments to `test_change`. ü§Ø
